```python




def proceed_infer(previous_result_collated: str, tool_intro_collated: str) -> str:
    prompt = f"Given the previous result:\n{previous_result_collated}\n\nAvailable tools:\n{tool_intro_collated}\n\nSelect an appropriate tool and argument to proceed with further refinement. Format:\nTool: <tool name>\nArg: <argument>\n You must include these 2 lines"
    return call_claude("You are an AI tool selector for iterative refinement.", prompt)


def collate_search_results(search_results):
    return "\n".join([f"Source: {source}\n{content}" for content, source in search_results])

def prepare_llm_tool_input(tools):
    return "\n".join([f"{name}: {description}" for name, description, _ in tools])

def agent_selection_output_parser(llm_output: str):
    lines = llm_output.split("\n")
    for line in lines:
        if "Tool: " in line:
            tool = line.split(": ")[1]
            continue
        if "Arg: " in line:
            argument = line.split(": ")[1]

    return tool, argument

def executor(tool, argument, tools=tools):
    for tool_name, _, function in tools:
        if tool_name == tool:
            return function(argument)
    return []

def reflect_on_results(collated_previous_results: str) -> str:
    prompt = f"Based on the following collated results:\n{collated_previous_results}\n\nDetermine if further refinement is needed or if a final answer has been reached. Respond concisely."
    return call_claude("You are an AI that evaluates iterative progress.", prompt)

def stepwise_summarization(results: list) -> str:
    prompt = f"Summarize the following stepwise results succinctly:\n{results}\n\nProvide a concise and coherent summary."
    return call_claude("You are an AI summarizer for iterative results.", prompt)

def prepare_llm_agent_prompt(base_prompt, tools):
    tool_info = prepare_llm_tool_input(tools)
    return f"{base_prompt}\n\nAvailable tools:\n{tool_info}\n\nDecide on the next step."

# Example usage:
system_prompt = "You are an AI that helps refine ideas iteratively."




user_prompt = "What is the most recent treatments on COVID-19"




base_prompt = f"The target is to solve the question: {user_prompt}\nDecide the next step logic and what problem should be solved next."

results = []
current_prompt = user_prompt






import time



STEP = 1


for step in range(STEP):


    if step > 0:
        base_prompt += f"\nPrevious steps indicate:\n{results[-1]['step_summary']}"
    
    current_prompt = prepare_llm_agent_prompt(base_prompt, tools)

    
    # Agent
    response = call_claude(system_prompt, current_prompt)
    tool_decision = proceed_infer(response, prepare_llm_tool_input(tools))
    tool, argument = agent_selection_output_parser(tool_decision)
    step_results = executor(tool, argument)
    collated_results = collate_search_results(step_results)
    
    # Reflection + reasoning
    step_conclusion = reflect_on_results(collated_results)
    step_summary = stepwise_summarization([response, collated_results, step_conclusion])
    
    stepwise_results = {
        "tool_decision": tool_decision,
        "tool_results": step_results,
        "inference": step_conclusion,
        "step_summary": step_summary
    }
    results.append(stepwise_results)
    



    time.sleep(20)


    print(f"Step {step+1}:\n{tool_decision}\nTool findings:\n{collated_results}\n{step_conclusion}\n{step_summary}\n")
    if "Final result reached" in step_conclusion:
        break

final_paper = stepwise_summarization([r["step_summary"] for r in results])
print("Final Conclusion Paper:\n", final_paper)
```
